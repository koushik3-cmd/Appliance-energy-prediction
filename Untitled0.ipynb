```python
# prompt: perform the above data frame and give the python code to show the linear regression between two columns

import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load the dataframe (assuming 'df' is already defined)

# Select two columns for linear regression
x = df['T_out']  # Replace with your desired x-column
y = df['Appliances']  # Replace with your desired y-column

# Add a constant term to the independent variable
x = sm.add_constant(x)

# Fit the linear regression model
model = sm.OLS(y, x).fit()

# Print the model summary
print(model.summary())

# Plot the regression line
plt.scatter(df['T_out'], df['Appliances'])
plt.plot(df['T_out'], model.predict(x), color='red')
plt.xlabel('T_out')  # Replace with your desired x-label
plt.ylabel('Appliances')  # Replace with your desired y-label
plt.title('Linear Regression')
plt.show()
```

```python
## Divide the columns based on type for clear column management

col_temp = ["T1","T2","T3","T4","T5","T6","T7","T8","T9"]

col_hum = ["RH_1","RH_2","RH_3","RH_4","RH_5","RH_6","RH_7","RH_8","RH_9"]

col_weather = ["T_out", "Tdewpoint","RH_out","Press_mm_hg",
                "Windspeed","Visibility"]

col_randoms = ["rv1", "rv2"]

col_target = ["Appliances"]
```

```python
# Seperate dependent and independent variables
feature_vars = df[col_temp + col_hum + col_weather + col_randoms]
target_vars = df[col_target]
```

```python
# Spliting dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature_vars, target_vars, test_size=0.20, random_state=0) # Assign the output to different variables
print (X_train.shape, X_test.shape) # Use the new variable names to access the train and test sets
```

```python
#linear regression model
from sklearn.linear_model import LinearRegression
lreg= LinearRegression()
#Spliting dataset into train and test
from sklearn.model_selection import train_test_split # Make sure to include this import
X_train, X_test, y_train, y_test = train_test_split(feature_vars, target_vars, test_size=0.20, random_state=0) # Assign the output to different variables
print (X_train.shape, X_test.shape) # Use the new variable names to access the train and test sets
lreg.fit(X_train, y_train) # Changed x_train to X_train
print(f'Train score for Linear regression model without using Kfold validation is: {lreg.score(X_train, y_train)}') # Changed x_train to X_train

from sklearn.model_selection import cross_val_score # Import the cross_val_score function
train_scores= cross_val_score(lreg,X_train,y_train, cv=10) # Changed x_train to X_train
print(f'Train score for Linear regression model using 10Fold cross validation is:  {train_scores.mean()}')
test_scores= cross_val_score(lreg,X_test,y_test, cv=10) # Changed x_test to X_test
print(f'Test score for Linear regression model using 10Fold cross validation is:  {test_scores.mean()}')
```

```python
mse= cross_val_score(lreg, X_train, y_train, scoring="neg_mean_squared_error", cv=10) # Changed x_train to X_train
print(f'RMSE for Linear regression model using 10Fold cross validation is: {mse.mean()}')
```
